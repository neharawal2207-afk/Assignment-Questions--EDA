Q11) Identify and remove outliers in the km_driven column using the IQR methodB

```python
import pandas as pd

# Load the CSV file from the provided raw GitHub URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA---1/51f80f61c2d635c5ea0021a7ead19193626efb10/BIKE%20DETAILS%20(1).csv"
df = pd.read_csv(url)

# Calculate the first quartile (Q1) and the third quartile (Q3) for km_driven
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
filtered_df = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Display the number of rows before and after removing outliers
rows_before = df.shape[0]
rows_after = filtered_df.shape[0]

print(f"Rows before removing outliers: {rows_before}")
print(f"Rows after removing outliers: {rows_after}")
```

Q13) What is the average depreciation in selling price based on the bike's age (current year - manufacturing
year)

```python
import pandas as pd
import datetime

# Load the CSV file from the provided raw GitHub URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA---1/51f80f61c2d635c5ea0021a7ead19193626efb10/BIKE%20DETAILS%20(1).csv"
df = pd.read_csv(url)

# Assuming the current year is 2025
current_year = 2025

# Calculate the age of each bike
df['age'] = current_year - df['year']

# Group by age and calculate the average selling price for each age
average_price_by_age = df.groupby('age')['selling_price'].mean()

# Check if age 0 exists in the grouped data
if 0 in average_price_by_age.index:
    average_price_new_bike = average_price_by_age[0]
    # Calculate the depreciation for each age
    depreciation = average_price_new_bike - average_price_by_age
    print("Depreciation from a new bike (age 0):\n", depreciation)
elif not average_price_by_age.empty:
    # If age 0 is not present, find the youngest bike in the dataset to use as a baseline
    youngest_age = average_price_by_age.index.min()
    youngest_bike_price = average_price_by_age[youngest_age]
    print(f"No bikes with 'age 0' (current year) found in the dataset. Using average price of age {youngest_age} as baseline: {youngest_bike_price:.2f}")

    # Calculate depreciation relative to the youngest available bike
    depreciation_from_youngest = youngest_bike_price - average_price_by_age
    print(f"\nDepreciation relative to bikes of age {youngest_age}:\n", depreciation_from_youngest)
else:
    print("No data available to calculate average prices by age.")



```

Q14)Which bike names are priced significantly above the average price for their manufacturing year?

```python
import pandas as pd

# Load the CSV file from the provided raw GitHub URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA---1/51f80f61c2d635c5ea0021a7ead19193626efb10/BIKE%20DETAILS%20(1).csv"
df = pd.read_csv(url)

# Calculate the average selling price for each manufacturing year
average_price_by_year = df.groupby('year')['selling_price'].mean()

# Merge the average price data back into the original dataframe
df = df.merge(average_price_by_year.rename('average_price'), left_on='year', right_index=True)

# Define a threshold to consider a price as significantly above average (e.g., 20% above the average price)
threshold = 1.2

# Identify bikes that are priced significantly above the average for their year
significantly_above_average = df[df['selling_price'] > threshold * df['average_price']]

# Extract the bike names that meet this criterion
bike_names_above_average = significantly_above_average['name'].unique()

print(bike_names_above_average)
```

```python
import pandas as pd
from collections import Counter
import re

# Load the CSV
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA---3/refs/heads/main/amazon.csv"
df = pd.read_csv(url)

# Display columns to check which column has product names
print(df.columns)

# Assuming 'product_name' is the column with product names
# Optional: use a popularity column if available, e.g., 'ratings_count'
popularity_col = 'ratings_count' if 'ratings_count' in df.columns else None

# Function to extract words
def extract_words(text):
    # Convert to lowercase, remove non-alphanumeric characters, split into words
    return re.findall(r'\b\w+\b', str(text).lower())

# Create a Counter for keywords
keyword_counter = Counter()

for idx, row in df.iterrows():
    words = extract_words(row['product_name'])
    if popularity_col:
        # If popularity exists, weight by it
        keyword_counter.update({word: row[popularity_col] for word in words})
    else:
        keyword_counter.update(words)

# Most common keywords
top_keywords = keyword_counter.most_common(20)
print("Top 20 Product Keywords:\n", top_keywords)
```

Q1) Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do
the needful?

```python
import pandas as pd

# Load the CSV from the new URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA-4/refs/heads/main/spotify.csv"
df = pd.read_csv(url)

# Check for null values
null_summary = df.isnull().sum()
print("Null Values Summary:\n", null_summary)

# Handle null values
for column in df.columns:
    if df[column].dtype == 'object':  # Categorical
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)
    else:  # Numeric
        median_value = df[column].median()
        df[column].fillna(median_value, inplace=True)

# Check for duplicates
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# Remove duplicate rows if any
if duplicate_rows > 0:
    df = df.drop_duplicates()
    print("Duplicate rows have been removed.")
```

Q9)Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV from raw GitHub URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA-4/main/spotify.csv"
df = pd.read_csv(url)

# Convert duration from milliseconds to minutes for easier interpretation
if 'Duration (ms)' in df.columns:
    df['duration_min'] = df['Duration (ms)'] / 60000

# Choose top 10 artists by number of tracks to make visualization readable
top_artists = df['Artist'].value_counts().head(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

# Box Plot
plt.figure(figsize=(12,6))
sns.boxplot(x='Artist', y='duration_min', data=df_top_artists, palette='Set2', hue='Artist', legend=False)
plt.title('Distribution of Track Durations Across Top 10 Artists (Box Plot)')
plt.xlabel('Artist')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.show()

# Violin Plot (alternative)
plt.figure(figsize=(12,6))
sns.violinplot(x='Artist', y='duration_min', data=df_top_artists, palette='Set3', hue='Artist', legend=False)
plt.title('Distribution of Track Durations Across Top 10 Artists (Violin Plot)')
plt.xlabel('Artist')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```

Q10  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV from raw GitHub URL
url = "https://raw.githubusercontent.com/neharawal2207-afk/EDA-4/main/spotify.csv"
df = pd.read_csv(url)

# Select top 10 artists by number of tracks for readability
top_artists = df['Artist'].value_counts().head(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

# Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(x='Artist', y='Popularity', data=df_top_artists, hue='Artist', palette='muted', legend=False)
plt.title('Distribution of Track Popularity Across Top 10 Artists (Violin Plot)')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()

# Swarm Plot (alternative)
plt.figure(figsize=(12,6))
sns.swarmplot(x='Artist', y='Popularity', data=df_top_artists, hue='Artist', size=3, palette='Set2', legend=False)
plt.title('Distribution of Track Popularity Across Top 10 Artists (Swarm Plot)')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```
